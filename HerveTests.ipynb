```python
import chromadb

# 1. On se connecte au dossier de données
# Le chemin /project/data/ est le chemin standard dans Workbench
client = chromadb.PersistentClient(path="/project/data/")

# 2. On récupère la collection
# Tape client.list_collections() pour vérifier le nom exact.
collection = client.get_collection(name="rag-chroma")

# 3. On récupère TOUT (les textes, les noms de fichiers et les IDs)
# On ne demande pas les 'embeddings' car ce sont des listes de chiffres illisibles
results = collection.get(include=['documents', 'metadatas'])

# 4. On affiche proprement
print(f"Nombre d'éléments dans la base : {len(results['ids'])}")
print("-" * 30)

for i in range(min(5, len(results['ids']))): # Affiche les 5 premiers pour tester
    source = results['metadatas'][i].get('source', 'Inconnue')
    texte = results['documents'][i][:200] # On coupe à 200 caractères pour la lisibilité
    print(f"SOURCE : {source}")
    print(f"EXTRAIT : {texte}...")
    print("-" * 10)
```
